```python
df["Species"]=df["Species"].replace({"Iris-setosa":1,"Iris-versicolor":2,"Iris-virginica":3}) 
```

```python
plt.hist(df['SepalLengthCm'], edgecolor='black')
plt.hist(df['SepalWidthCm'], edgecolor='blue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Sepal Length and Sepal Width Distributions')
```

```python
plt.hist(df['PetalLengthCm'], edgecolor='black')
plt.hist(df['PetalWidthCm'], edgecolor='blue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Sepal Length and Sepal Width Distributions')
```
