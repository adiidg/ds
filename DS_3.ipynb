```python
import pandas as pd
df=pd.read_csv('adult.csv')
df
```

```python
df.head()
```

```python
mean=df.groupby('sex')['age'].mean()
```

```python
mean
```

```python
median=df.groupby('race')['age'].median()
median
```

```python
min=df.groupby('race')['age'].min()
min
```

```python
max=df.groupby('race')['age'].max()
max
```

```python
std=df.groupby('race')['age'].std()
std
```

```python
grouped_stats=df.groupby('sex')['age'].agg(['mean','median','std','min','max'])
grouped_stats
```

```python
grouped_stats=df.groupby('sex')[['age','hours.per.week']].describe()
grouped_stats
```

```python
import pandas as pd
iris=pd.read_csv('iris.csv')
iris.head()
```

```python
setosa=iris[iris['species']=='setosa']
virginica=iris[iris['species']=='virginica']
versicolor=iris[iris['species']=='versicolor']

setosa_stats=setosa.describe()
virginica_stats=virginica.describe()
versicolor_stats=versicolor.describe()
```

```python
print("summary for setosa")
print(setosa_stats)
print("summary for virginica")
print(virginica_stats)
print("summary for versicolor")
print(versicolor_stats)
```
