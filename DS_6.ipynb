```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score, confusion_matrix
```

```python
df=pd.read_csv('iris.csv')
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True,cbar=False)
plt.xlabel('predicted_label')
plt.ylabel('actual_label')
```

```python
print('\nconfusion matrix : \n',cm) 
print('\nf1 score : ',f1_score(y_test,y_pred,average='macro'))  #Macro averaging is helpful when the class distribution is imbalanced because it ensures each class contributes equally to the final metric.
print('\naccuracy score : ',accuracy_score(y_test,y_pred))
print('\nrecall score : ',recall_score(y_test,y_pred,average='macro'))
print('\nprecision score : ',precision_score(y_test,y_pred,average='macro'))
```
