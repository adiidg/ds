```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='sex',hue='survived',data=df,palette='Set1')
plt.title('Survival count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='survived',labels=['no','yes'])
plt.tight_layout()
plt.plot()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='class',hue='survived',data=df,palette='Set1')
plt.title('Survival count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='survived',labels=['no','yes'])
plt.tight_layout()
plt.plot()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(x='age',data=df,bins=30,kde=True,color='red')
plt.title('Age distribution by no. of passengers')
plt.xlabel('Age')
plt.ylabel('no. of passengers')
plt.legend(title='survived',labels=['no','yes'])
plt.grid('on')
plt.tight_layout()
plt.plot()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(x='fare',data=df,bins=30,kde=True,color='red')
plt.title('Fare distribution by no. of passengers')
plt.xlabel('Fare')
plt.ylabel('no. of passengers')
plt.legend(title='survived',labels=['no','yes'])
plt.grid('on')
plt.tight_layout()
plt.plot()
```
