```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
df[['sex','age','survived']].head()
```

```python
sns.boxplot(x='sex',y='age',data=df)
```

```python
df['survived_label']=df['survived'].map({0:'Not Survived',1:'Survived'})
```

```python
custom_palette = {'Not Survived':'red','Survived':'green'}
```

```python
sns.boxplot(x='sex',y='age',hue='survived_label',data=df,palette=custom_palette)
```
